Importação das Bibliotecas

```python
# Manipulação de dados
import pandas as pd
import numpy as np

# Visualização de dados
import plotly.express as px

# Manipulação de arquivos (download)
from google.colab import files
```

Importação dos arquivos .csv

```python
df_cor_raca = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SIN 323 - Projeto 02/csv/CID-10_COR_RAÇA.csv', encoding='utf-8', sep=';')
df_idade = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SIN 323 - Projeto 02/csv/CID-10_IDADE.csv', encoding='utf-8', sep=';')
df_regiao = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SIN 323 - Projeto 02/csv/CID-10_REGIAO.csv', encoding='utf-8', sep=';')
df_sexo = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SIN 323 - Projeto 02/csv/CID-10_SEXO.csv', encoding='utf-8', sep=';')
df_unidade = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/SIN 323 - Projeto 02/csv/CID-10_UNIDADE.csv', encoding='utf-8', sep=';')
```

Substituindo os valores nulos ("-") por 0

```python
df_cor_raca.replace("-", 0, inplace=True)
df_idade.replace("-", 0, inplace=True)
df_regiao.replace("-", 0, inplace=True)
df_sexo.replace("-", 0, inplace=True)
df_unidade.replace("-", 0, inplace=True)
```

Convertendo os valores para INT

```python
for col in df_cor_raca.columns[1:]:
    df_cor_raca[col] = df_cor_raca[col].astype(int)

for col in df_idade.columns[1:]:
    df_idade[col] = df_idade[col].astype(int)

for col in df_regiao.columns[1:]:
    df_regiao[col] = df_regiao[col].astype(int)

for col in df_sexo.columns[1:]:
    df_sexo[col] = df_sexo[col].astype(int)

for col in df_unidade.columns[1:]:
    df_unidade[col] = df_unidade[col].astype(int)
```

Gráficos CID-10_COR_RACA.csv

```python
for raca in df_cor_raca.columns[1:]:
  fig_cor_raca = px.bar(df_cor_raca,
                        x = 'Capítulo CID-10',
                        y = raca,
                        labels = {'Capítulo CID-10': 'Capítulo CID-10',
                                  raca: 'Óbitos'},
                        title = f"Quantidade de óbitos registrados por CID-10 em pessoas de cor/raça {raca}")
  fig_cor_raca.show()
  fig_cor_raca.write_html(f"fig_cor_raca_{raca}.html")
```

Gráficos CID-10_IDADE.csv

```python
for idade in df_idade.columns[1:]:
  fig_idade = px.bar(df_idade,
                     x = 'Capítulo CID-10',
                     y = idade,
                     labels = {'Capítulo CID-10': 'Capítulo CID-10',
                               idade: 'Óbitos'},
                     title = f"Quantidade de óbitos registrados por CID-10 em pessoas de idade {idade}")
  fig_idade.show()
  fig_idade.write_html(f"fig_idade_{idade}.html")
```

Gráficos CID-10_REGIAO.csv

```python
for regiao in df_regiao.columns[1:]:
  fig_regiao = px.bar(df_regiao,
                      x = 'Capítulo CID-10',
                      y = regiao,
                      labels = {'Capítulo CID-10': 'Capítulo CID-10',
                               regiao: 'Óbitos'},
                     title = f"Quantidade de óbitos registrados por CID-10 em {regiao}")
  fig_regiao.show()
  fig_regiao.write_html(f"fig_regiao_{regiao}.html")
```

Gráficos CID-10_SEXO.csv

```python
for sexo in df_sexo.columns[1:]:
  fig_sexo = px.bar(df_sexo,
                      x = 'Capítulo CID-10',
                      y = sexo,
                      labels = {'Capítulo CID-10': 'Capítulo CID-10',
                               sexo: 'Óbitos'},
                     title = f"Quantidade de óbitos registrados por CID-10 em pessoas do sexo {sexo}")
  fig_sexo.show()
  fig_sexo.write_html(f"fig_sexo_{sexo}.html")
```

Gráficos CID-10_UNIDADE.csv

```python
for unidade in df_unidade.columns[1:]:
  fig_unidade = px.bar(df_unidade,
                      x = 'Capítulo CID-10',
                      y = unidade,
                      labels = {'Capítulo CID-10': 'Capítulo CID-10',
                               unidade: 'Óbitos'},
                     title = f"Quantidade de óbitos registrados por CID-10 em {unidade}")
  fig_unidade.show()
  fig_unidade.write_html(f"fig_unidade_{unidade}.html")
```
