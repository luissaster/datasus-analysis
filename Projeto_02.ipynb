```python
anos = [2019, 2020, 2021, 2022, 2023]
obitos = [1349801, 1556824, 1832649, 1544266, 1465610]

plt.figure(figsize=(10, 6))
plt.plot(anos, obitos, marker='o', linestyle='-', color='b', label='Óbitos')
plt.title('Evolução do Número de Óbitos por Ano', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Número de Óbitos', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.xticks(anos)
plt.legend()

for i, valor in enumerate(obitos):
    plt.text(anos[i], obitos[i] + 20000, f'{valor:,}'.replace(',', '.'),
             ha='center', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

```python
anos = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]).reshape(-1, 1)
mortos = np.array([1136947, 1170498, 1181166, 1210474, 1227039, 1264175, 1309774, 1312663, 1316719, 1349801, 1556824, 1832649, 1544266, 1465610])

modelo = LinearRegression()
modelo.fit(anos, mortos)

anos_futuros = np.array([2024, 2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033]).reshape(-1, 1)
previsoes = modelo.predict(anos_futuros)

mortos_previstos = modelo.predict(anos)

plt.scatter(anos, mortos, color='blue', label='Dados Históricos')

plt.plot(anos, mortos_previstos, color='red', label='Regressão Linear')

plt.scatter(anos_futuros, previsoes, color='green', label='Previsões Futuras')

plt.xlabel('Ano')
plt.ylabel('Quantidade de Óbitos')
plt.title('Regressão Linear e Previsão de Óbitos no Brasil')
plt.legend()

plt.grid(True)
plt.show()

for ano, previsao in zip(anos_futuros, previsoes):
    print(f"Ano: {ano[0]}, Previsão de Óbitos: {previsao:.0f}")

r = modelo.score(anos, mortos)
print(f"Coeficiente de Determinação (R^2): {r:.4f}")
```
